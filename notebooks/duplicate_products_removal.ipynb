```python
import pandas as pd
products = pd.read_csv('./data/products.csv', sep = '\t')
ratings = pd.read_csv('./data/ratings.csv', sep='\t')
links = pd.read_csv('./data/product_links.csv', sep='\t')
```

```python
print('Number of links: ', len(links))
print('Number of unique links', len(links.links.unique()))
```

```python
duplicated_links = links[links.duplicated(keep='first')]
duplicated_links
```

```python
DUPLICATED_IDs = [623, 663, 745, 759, 771, 773, 781, 786]
products[products.id.isin(DUPLICATED_IDs)]
```

```python
products = products.drop_duplicates(subset=['id'])
print('Number of scraped products: ', len(products))
```

```python
products_no_duplicates = products[~products.id.isin(DUPLICATED_IDs)].copy()
products_no_duplicates.reset_index(inplace=True, drop=True)
products_no_duplicates['new_id'] = products_no_duplicates.index
products_no_duplicates
```

```python
ratings
```

```python
ratings_index60 = ratings[ratings.product_id == 60].drop_duplicates()
ratings_index60
```

```python
ratings_no_60 = ratings[ratings.product_id != 60].copy()
ratings_no_duplicates = pd.concat([ratings_no_60, ratings_index60]).reset_index(drop=True)
ratings_no_duplicates
```

```python
ratings_no_duplicates = ratings_no_duplicates.query('product_id not in @DUPLICATED_IDs').copy()
ratings_no_duplicates.reset_index(drop=True, inplace=True)
ratings_no_duplicates.product_id.max()
```

```python
ratings_no_duplicates['product_id'] = ratings_no_duplicates['product_id'].replace(
    products_no_duplicates.set_index('id')['new_id'])
ratings_no_duplicates['product_id'].max()
```

```python
products_no_duplicates['id'] = products_no_duplicates['new_id']
```

```python
del products_no_duplicates['new_id']
products_no_duplicates
```

```python
products_no_duplicates.to_csv('./data/products_nd.csv', sep='\t', index=False)
ratings_no_duplicates.to_csv('./data/ratings_nd.csv', sep='\t', index=False)
```
